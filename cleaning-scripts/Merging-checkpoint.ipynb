## Retention

```python
# load in the data
filename = 'Retention_2011-12_National.csv'
f = pd.read_csv(filename)

# create temporary output dataframe
Output = pd.DataFrame(columns=['Description','Grade','RaceEthnicity','Gender','Disabilities','IDEA','s504','LEP','ELL','Credentials','Training',
                               'Security','Country','Region','State','County','StartYear','EndYear','Value','Percent','Source'], index = range(len(f)))

# set the description, start year, end year (based on the *dataset* that we are actively merging)
Output.Description = ['Students Retained']*len(f)
StartYear, EndYear = 2011, 2012
Output.StartYear = [StartYear]*len(f)
Output.EndYear = [EndYear]*len(f)

# set other columns that are easy to set 
Output.Grade = f.Grade
Output.Value = f.Number
Output.Percent = f.Percent
Output.Source = filename

# others that are harder to set
# first, recode the gender variable (it sounds better as 'all')
gender = np.where(f.Gender=='total','all', f.Gender) 
Output.Gender = gender

# specify the region based on the StatesToRegions dictionary in the 2nd cell
Output.Country = ['US']*len(f)
state = np.where(f.State=='KA','KS',f.State)
Output.State = state
Output.Region = [StatesToRegions[s] for s in state]

# recode the 'Attribute' field from the Retention dataset to separate disability from the race data
# so these columns are T/F values
Output.Disabilities = (np.array(f.Attribute) =='Students with Disabilities (all)') | (np.array(f.Attribute)=='Students With Disabilities Served Only Under Section 504') | (np.array(f.Attribute)=='Students With Disabilities Served Under IDEA') | (np.array(f.Attribute)=='English Language Learners with Disabilities')
Output.IDEA = np.array(f.Attribute) == 'Students With Disabilities Served Under IDEA'
Output.s504 = np.array(f.Attribute) == 'Students With Disabilities Served Only Under Section 504'
Output.ELL = np.array(f.Attribute) == 'English Language Learners'

# now that we have saved IDEA/s504/ELL data, recode the Attribute field to save the rest under RaceEthnicity
attribute = np.where(f.Attribute=='Students With Disabilities Served Only Under Section 504','all',f.Attribute)
attribute = np.where(attribute=='Students With Disabilities Served Under IDEA', 'all', attribute)
attribute = np.where(attribute=='English Language Learners', 'all', attribute)
attribute = np.where(attribute=='Total Students', 'all',attribute)
Output.RaceEthnicity = attribute

Output.head(5)
final_output = pd.concat([final_output, Output])
print(len(final_output),'rows')
final_output.head(5)
```

```python
# load in the data
filename = 'Discipline_2011-2012_National.csv'
f = pd.read_csv(filename, encoding='latin')

da = np.array(f['Disciplinary Action'],dtype=str)

newda, gender = [], []
for s in da:
    if s.endswith('on females') or s.endswith('on Females'):
        newda.append(s[0:-11])
        gender.append('female')
    elif s.endswith('for females') or s.endswith('for Females') or s.endswith('for\xa0females'):
        newda.append(s[0:-12])
        gender.append('female')
    elif s.endswith('on males') or s.endswith('on Males'):
        newda.append(s[0:-9])
        gender.append('male')
    elif s.endswith('for males') or s.endswith('for Males') or s.endswith('for\xa0males'):
        newda.append(s[0:-10])
        gender.append('male')
    elif s.endswith('total') or s.endswith('Total'):
        newda.append(s[0:-6])
        gender.append('all')
        
newda = np.array(newda)

# create temporary output dataframe
Output = pd.DataFrame(columns=['Description','Grade','RaceEthnicity','Gender','Disabilities','IDEA','s504','LEP','ELL','Credentials','Training',
                               'Security','Country','Region','State','County','StartYear','EndYear','Value','Percent','Source'], index = range(len(f)))


# set the description, start year, end year (based on the *dataset* that we are actively merging)
Output.Description = newda
StartYear, EndYear = 2011, 2012
Output.StartYear = [StartYear]*len(f)
Output.EndYear = [EndYear]*len(f)

# set other columns that are easy to set 
#Output.Grade = Participation.Grade # no grade data here
Output.Value = f.Number
Output.Percent = f.Percent
Output.Source = filename

# others that are harder to set
Output.Gender = gender

# specify the region based on the StatesToRegions dictionary in the 2nd cell
Output.Country = ['US']*len(f)
state = np.where(f.State=='KA','KS',f.State)
Output.State = state
Output.Region = [StatesToRegions[s] for s in state]

# recode the 'Attribute' field from the Retention dataset to separate disability from the race data
# so these columns are T/F values
Output.Disabilities = (np.array(f.Attribute) =='Students with Disabilities (all)') | (np.array(f.Attribute)=='Students with Disabilities (504)') | (np.array(f.Attribute)=='Students with Disabilities (IDEA)') | (np.array(f.Attribute)=='English Language Learners with Disabilities')
Output.IDEA = np.array(f.Attribute) == 'Students with Disabilities (IDEA)'
Output.s504 = np.array(f.Attribute) == 'Students with Disabilities (504)' 
Output.ELL = (np.array(f.Attribute) == 'English Language Learners') | (np.array(f.Attribute)=='English Language Learners with Disabilities')

# now that we have saved IDEA/s504/ELL data, recode the Attribute field to save the rest under RaceEthnicity
# attribute = np.where(f.Attribute=='Students With Disabilities Served Only Under Section 504','all',f.Attribute)
attribute = np.where(f.Attribute=='Students with Disabilities (all)','all',f.Attribute)
attribute = np.where(attribute=='Students with Disabilities (504)','all',attribute)
attribute = np.where(attribute=='Students with Disabilities (IDEA)','all',attribute)
attribute = np.where(attribute=='English Language Learners with Disabilities','all',attribute)
attribute = np.where(attribute=='English Language Learners', 'all', attribute)



#### need to fix these attributes... 

Output.RaceEthnicity = attribute

Output.head(5)
final_output = pd.concat([final_output, Output])
print(len(final_output),'rows')
final_output.head(5)
```
